This data analysis uses a data set of tweets captured from April 3rd, 2013 to April 4th, 2013. With this data I hope to determine the overall feelings towards climate change. I will use a number of tools to determine what the conversation is geared towards and sentiment analysis to understand the overall tone of the conversation. 

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import warnings
import nltk
from nltk import tokenize as tk
from nltk import FreqDist
from nltk.corpus import stopwords
import string
from matplotlib import pyplot as plt
```

```python
clim_uniq = pd.read_csv('/Volumes/Drive2/EMAT22110_Notebooks/ClimateChange_Apr3-Apr4_2013_fixed.csv')
clim_uniq.keys()
```

The data is in a CSV file the contains the above information for tweet that was collected.

```python
en_only = clim_uniq.loc[clim_uniq['Language'] == 'en']
en_only
```

```python
en_only['text_lower'] = [x.lower() for x in en_only['Full Text']]
en_only['text_lower'].head(20)
```

Here I made all of the text lowercase in order to run a word frequency analysis, that way the text is normalized and can be adequately measured.

```python
en_only['Full Text'][224]
```

```python
en_only['tokens'] = [tk.word_tokenize(x) for x in en_only['text_lower']]
en_only['tokens'].head(5)
```

```python
en_only['tokens'][2]
```

I then tokenized the text so the I have a list of every word in every tweet.

```python
len(stopwords.words('english'))
```

```python
mystopwords = stopwords.words('english')
mystopwords.append("n't")
mystopwords.append("'s")
mystopwords.append("...")
mystopwords.append("``")
mystopwords.append("''")
mystopwords.append("’")
mystopwords.append("http")
mystopwords.append("rt")
len(mystopwords)
```

Here I have implemented the use of stop words to weed out any common words that would appear frequently but have little to no bearing on the conversation. I added in some of my own words here that appeared frequently but I determined are not relevant to the data. 

```python
filtered = []
for i in en_only['tokens']:
    filtered.append([word for word in i if word not in mystopwords])
    
en_only['filtered'] = filtered
```

```python
en_only['filtered'][2]
```

```python
string.punctuation
```

```python
no_punct = []
for i in en_only['filtered']:
    no_punct.append([word for word in i if word not in string.punctuation])
    
en_only['no_punct'] = no_punct
en_only['no_punct'][2]
```

```python
freq1 = [FreqDist(x) for x in en_only['no_punct']]
en_only['freqs1'] = freq1
en_only['freqs1']
```

```python
from collections import Counter
```

```python
A = Counter(en_only['freqs1'][0])
B = Counter(en_only['freqs1'][1])
A + B
```

```python
en_only['f1_counts'] = [Counter(x) for x in en_only['freqs1']]
en_only['f1_counts'][0]
```

```python
wordcounts = Counter()
for i in en_only['f1_counts']:
    wordcounts += i
```

```python
len(wordcounts)
```

```python
type(wordcounts)
```

```python
dict(wordcounts)
```

```python
counts = pd.DataFrame.from_dict(dict(wordcounts), orient='index')
counts
```

```python
counts = counts.rename(columns = {0: 'counts'})
counts
```

Here is the final list for the word frequency analysis, below is the list of word frequency in descending order of frequency. 

```python
count_sort = counts.sort_values('counts', ascending = False)[:40]
count_sort
```

```python
fig, ax0 = plt.subplots(figsize=(15,10))
ax0.bar(count_sort.index, count_sort['counts'])
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

After stemming the words to increase accuracy the count is relative unchanged with the most frequent words being 'climate', 'change' and 'climatechange' which is presumably a hashtag. 

```python
stemmer = nltk.stem.SnowballStemmer("english")
stems = []
for i in en_only['no_punct']:
    stems.append([stemmer.stem(word) for word in i])
    
en_only['stemmed'] = stems
```

```python
freq2 = [FreqDist(x) for x in en_only['stemmed']]
en_only['freqs2'] = freq2
en_only['freqs2']
```

```python
en_only['f2_counts'] = [Counter(x) for x in en_only['freqs2']]
en_only['f2_counts'][0]
```

```python
wordcounts = Counter()
for i in en_only['f2_counts']:
    wordcounts += i
```

```python
counts2 = pd.DataFrame.from_dict(dict(wordcounts), orient='index')
counts2
```

```python
counts2 = counts2.rename(columns = {0: 'counts'})
counts2
```

```python
count_sort2 = counts2.sort_values('counts', ascending = False)[:20]
count_sort2
```

```python
fig, ax1 = plt.subplots(figsize=(10,10))
ax1.bar(count_sort2.index, count_sort2['counts'])
plt.xticks(rotation=90, fontsize = 'medium')
plt.show()
```

```python
analyser = SentimentIntensityAnalyzer()
warnings.filterwarnings('ignore')
```

```python
print(en_only['Full Text'][3])
print(analyser.polarity_scores(en_only['Full Text'][3]))
```

```python
en_only['vader_scores'] = [analyser.polarity_scores(x) for x in en_only['Full Text']]
en_only['vader_scores'].head(20)
```

```python
en_only['pos'] = [x['pos'] for x in en_only['vader_scores']]
en_only['pos']
```

```python
fig, ax0 = plt.subplots(figsize = (10,10))
ax0.plot(en_only['pos'])
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

```python
type(en_only['Date'][0])
```

```python
from datetime import datetime as dt
```

```python
en_only['Datetime'] = [dt.strptime(x, '%Y-%m-%d %H:%M:%S.%f') for x in en_only['Date']]
print(type(en_only['Datetime'][0]))
print(en_only['Datetime'][0])
```

```python
fig, ax1 = plt.subplots(figsize = (10,10))
ax1.scatter(en_only['Datetime'], en_only['pos'])
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```
